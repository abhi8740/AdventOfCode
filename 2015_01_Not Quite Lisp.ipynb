```python
def calculate_floor(input_data):
    floor = 0
    for letter in input_data:
        if letter == '(':
            floor+=1    # An opening parenthesis, (, means he should go up one floor,
        else:
            floor-=1    # A closing parenthesis, ), means he should go down one floor.
    return floor
```
